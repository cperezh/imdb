```python
import pandas as pd
import psycopg2
```

```python
conn = psycopg2.connect(
    host="localhost",
    database="imdb",
    user="postgres",
    password="postgres")
```

```python
query = '''
select endyear , count(*)
from ods.basics
group by endyear
'''
```

```python
df = pd.read_sql_query(query,con=conn)
```
